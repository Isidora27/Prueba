````python
## Obtener información de los números de ortología de forma iterativa
## archivo que va a guardar la info de las reacciones
with open('reacciones_clase_2021.txt','w') as out,\
    open('secuencias_aa_2021.fasta','w') as out2: ## archivo para guardar secuencias de aminoacidos
    eq_list = [] ## crea una lista vacía para guardar las ecuaciones de las reacciones
    for item in ort_num: # iteramos para cada número de esta lista
        ort = k.get('ko:'+item) ## Obtener información de cada número de ortología 
        ort_parsed = k.parse(ort) ## Obtener la información en un diccionario 
        def_ort = ort_parsed['NAME'] ## Obtener la definición del número de ortología 
        dblink = ort_parsed['DBLINKS'] ## obtener la información guardada con la llave DBLINKS
        for obj in dblink.keys():
            if obj == 'RN': ## Obtener el número de reacción asociado a cada número
                rns = dblink[obj].split(' ') ## Divide los objetos presentes en dblink[ob] por espacio
                for r in rns:
                    if r in  rn_num:
                         r_inf = k.get('rn:'+r) ## Obtener información de cada reacción 
                         r_inf_parsed = k.parse(r_inf) ## Ordenar la información en un diccionario
                         paths = r_inf_parsed['PATHWAY'] ## Obtener rutas metabólicas asociadas
                         defi = r_inf_parsed['DEFINITION'] ## Obtener ecuación con nombre de compuesto
                         eq = r_inf_parsed['EQUATION'] ## Obtener ecuación con código de compuestos de KEGG 
                         eq_list.append(eq) ## guarda las ecuaciones en una lista
                         enzyme = r_inf_parsed['ENZYME'] ## Obtener EC number
        genes = ort_parsed['GENES'] ## Obtener los genes asociados con cada número de ortología
        print(genes)
        g_key = genes.keys() ## Obtener las llaves de acceso en los genes (corresponden a las abb. de los mo en mayúsculas)
        print(g_key)
        for g in g_key:
            if g == 'SLE':
                g_sle = genes[g].split(' ') ## genes[g] recupera los genes de SLE, split(' ') separa los genes si hay mas de
                print(g_sle)
                for g_2 in g_sle:
                    print(g_2)
                    g_aa = k.get('sle:'+g_2.split('(')[0]+'/aaseq') 
                    #g_aa = k.get('sle:'+g_2+'/aaseq') ## Obtener las secuencias de aminoácidos
                    out2.write(g_aa+'\n') ## guardar las secuencias de aa en n archivo
        out.write(str(item)+'\t'+str(def_ort)+'\t'+str(r)+'\t'+str(g_sle)+'\t'+str(defi)+'\t'+str(eq)+'\t'+str(enzyme)+'\t'+str(paths)+'\n') ## el archivo se puede crear aquí ya que para cada número



````
